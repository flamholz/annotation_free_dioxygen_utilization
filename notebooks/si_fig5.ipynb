```python
import matplotlib.pyplot as plt
from aerobot.utils import RESULTS_PATH, DATA_PATH, load_results_dict, FIGURES_PATH
from aerobot.plot import plot_configure_mpl, plot_confusion_matrix
import pandas as pd
from matplotlib.gridspec import GridSpec
import numpy as np
import os
from typing import List, Dict, Tuple
%load_ext autoreload 
%autoreload 2 

plot_configure_mpl() # Make height same as width for a square confusion matrix. 
```

```python
rna16s_results = load_results_dict(os.path.join(RESULTS_PATH, 'train_linear_embedding_rna16s_ternary.json'))
```

# SI Figure 5

```python
def plot_si_figure_5(rna16s_results:Dict, fig:plt.Figure=None, grid:GridSpec=None):

    ax = fig.add_subplot()
    plot_confusion_matrix(rna16s_results, ax=ax)
    ax.set_xlabel('predicted label')
    ax.set_ylabel('true label')
    test_acc = 100 * np.round(rna16s_results['test_acc'], 2)
    ax.set_title(f'16S classifier validation accuracy ({test_acc}%)')
```

```python
fig = plt.figure(figsize=(2.5, 2.5))
plot_si_figure_5(rna16s_results, fig=fig)

plt.savefig(os.path.join(FIGURES_PATH, 'si_fig5.png'), dpi=300, facecolor='w', bbox_inches='tight')
plt.show()
```
